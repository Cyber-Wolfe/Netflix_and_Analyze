```python
%matplotlib inline
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
Score_df = pd.read_csv("../../csv_file/Netflix_data_ML.csv")
Genre_df = pd.read_csv("../../csv_file/Netflix_Dataset_Genre_Type.csv")
```

```python
Score_df.head()
```

```python
Genre_df.head()
```

```python
Score_Genre_df = pd.merge(Score_df, Genre_df, on = 'Key')
Score_Genre_df = Score_Genre_df.set_index('Key')
Score_Genre_df.head()
```

```python
Score_Genre_df.rename(columns = 
                      {'Hidden Gem Score':'Hidden_Gem_Score', 
                       'IMDb Score':'IMDb_Score', 
                       'Rotten Tomatoes Score':'Rotten_Tomatoes_Score', 
                       'Metacritic Score':'Metacritic_Score'}, 
                      inplace = True)
Score_Genre_df.head()
```

```python
Score_Genre_df["IMDb_Score"] = 10 * Score_Genre_df["IMDb_Score"]
Score_Genre_df["Hidden_Gem_Score"] = 10 * Score_Genre_df["Hidden_Gem_Score"]
Score_Genre_df.head()
```

```python
file_path = "../../csv_file/Score_Genre.csv"
Score_Genre_df.to_csv(file_path, index=True)
```
