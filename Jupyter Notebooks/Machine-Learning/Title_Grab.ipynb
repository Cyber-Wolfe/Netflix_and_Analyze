```python
%matplotlib inline
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
ScoreGenreEarnings = pd.read_csv("../../csv_file/Netflix_data_ML.csv")
Titles = pd.read_excel('../../Excel Sheets/Netflix Dataset Latest 2021 Filtered - Data_Ready_for_DF.xlsx', engine='openpyxl')
Box_Office_df = pd.read_csv("../../csv_file/key_Boxoffice.csv")
```

```python
Score_Genre_Earnings_Titles_df = pd.merge(ScoreGenreEarnings, Titles, on = 'Key')
Score_Genre_Earnings_Titles_df = Score_Genre_Earnings_Titles_df.set_index('Key')
Score_Genre_Earnings_Titles_df
```

```python
Box_Office_df.rename(columns = 
                      {'Title':'Box_Office_Earnings'}, 
                      inplace = True)
Score_Genre_Earnings_Titles_df.rename(columns = 
                      {'Hidden Gem Score':'Hidden_Gem_Score', 
                       'IMDb Score':'IMDb_Score', 
                       'Rotten Tomatoes Score':'Rotten_Tomatoes_Score', 
                       'Metacritic Score':'Metacritic_Score'}, 
                      inplace = True)
Score_Genre_Earnings_Titles_df
```

```python
Score_Genre_Earnings_Titles_df = pd.merge(Score_Genre_Earnings_Titles_df, Box_Office_df, on = 'Key')
Score_Genre_Earnings_Titles_df = Score_Genre_Earnings_Titles_df.set_index('Key')
Score_Genre_Earnings_Titles_df
```

```python
Score_Genre_Earnings_Titles_df = Score_Genre_Earnings_Titles_df.drop(['Series or Movie'], axis=1)
```

```python
for col in Score_Genre_Earnings_Titles_df.columns:
    print(col)
```

```python
Score_Genre_Earnings_Titles_df["IMDb_Score"] = 10 * Score_Genre_Earnings_Titles_df["IMDb_Score"]
Score_Genre_Earnings_Titles_df["Hidden_Gem_Score"] = 10 * Score_Genre_Earnings_Titles_df["Hidden_Gem_Score"]
```

```python
Score_Genre_Earnings_Titles_df = Score_Genre_Earnings_Titles_df[Score_Genre_Earnings_Titles_df['Box_Office_Earnings']!=0]
```
